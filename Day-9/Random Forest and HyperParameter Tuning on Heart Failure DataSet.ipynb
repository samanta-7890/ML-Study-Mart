Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart failure.csv')
df.head()
```

Info of data

```python
df.info()
```

Statistical Info of data

```python
df.describe()
```

Renaming column DEATH_EVENT as death

```python
df = df.rename({'DEATH_EVENT':'death'}, axis=1)
df.head()
```

Saving or downloading current csv file in local device

```python
df.to_csv('new.csv')
```

Data Visualization

```python
import seaborn as sns
```

Separating Features & Target Values

```python
x = df.drop('death', axis=1)
y = df['death']
```

```python
x.head()
```

Splitting the data as 70:30 ratio

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x_test.shape
```

```python
y_test.shape
```

# Apply Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

Training the data

```python
dt.fit(x_train, y_train)
```

```python
pred = dt.predict(x_test)
pred
```

Actual value of y_test

```python
np.array(y_test)
```

Evaluating DecisionTreeClassifier in terms of score

```python
dt.score(x_test, y_test)
```

Evaluating the ML model in terms of confusion matrix 

```python
from sklearn.metrics import classification_report, confusion_matrix, plot_roc_curve
```

```python
print(classification_report(y_test, pred))
```

```python
cnf = confusion_matrix(y_test, pred)
cnf
```

```python
sns.heatmap(cnf, annot=True)
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(x_train, y_train)
```

```python
rfc.score(x_test, y_test)
```

# HyperParameter Tuning Using Randomized SearchCV

```python
n_estimators = np.random.randint(20, 195, 25)
n_estimators
```

```python
criterion=["gini", "entropy"]
max_depth= np.random.randint(1,13,20)
min_samples_split=[2,3,4]
max_features=["sqrt", "log2"]
```

```python
# grids
ids = {
    'n_estimators':n_estimators,
    'criterion': criterion,
    'max_depth': max_depth,
    'min_samples_split': min_samples_split,
    'max_features': max_features
}
```

```python
print(ids)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rcv = RandomizedSearchCV(rfc, ids, n_iter=200, cv=3, random_state = 2)
```

```python
rcv.fit(x_train, y_train)
```

```python
rcv.cv_results_
```

```python
cv_result = pd.DataFrame(rcv.cv_results_)
```

```python
cv_result
```

```python
rcv.best_params_
```

```python
rcv.best_score_
```

```python
rcv.score(x_test, y_test)
```

```python
rcv.best_estimator_
```
