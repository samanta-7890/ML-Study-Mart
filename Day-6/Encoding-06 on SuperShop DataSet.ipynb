# One Hot Encoder Using For Loop

```python
df12.columns
```

```python
for col in df12.columns:
    if df12[col].dtype != np.number:
        one = pd.get_dummies(df12[col],prefix=col,drop_first=True)
        df12 = pd.concat([df12,one],axis=1).drop(col,axis=1)
        
```
