# Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

# Read Data

```python
df = pd.read_csv('supershops.csv')
df.head()
```

```python
df.shape
```

# Info of data

```python
df.info()
```

# Statistical Info Of Data

```python
df.describe()
```

# Check Null Value

```python
df.isnull().sum()
```

# Handling Null Value

```python
df.fillna(df['Transport'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

# Drop Duplicate Rows

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

# Separating Feature and Target

```python
x = df.drop('Profit', axis=1)
y = df['Profit']
```

```python
x.head()
```

```python
y.head()
```

# Data Preprocessing

```python
# Apply One Hot Encoding on 'Area' column
one = pd.get_dummies(x['Area'],prefix='Area',drop_first=True)
x = pd.concat([x,one],axis=1).drop('Area',axis=1)
```

```python
x.head()
```

# LinearRegression

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
y_pred = lr.predict(x)
y_pred
```

```python
# accuracy score
lr.score(x,y)
```

```python
# intercepts
c = lr.intercept_
c
```

```python
# coefficients
m = lr.coef_
m
```

# Prediction Using model object

```python
y_pred_1 = lr.predict([[142107.34,91391.77,366168.42,0,1]])
```

```python
y_pred_1
```

# Prediction Using Coefficient and Intercept

```python
# coefficients assigning
m1 = m[0]
m2 = m[1]
m3 = m[2]
m4 = m[3]
m5 = m[4]

#intercepts = c

#features
x1,x2,x3,x4,x5 = 142107.34,91391.77,366168.42,0,1

y_pred_2 = m1 * x1 + m2 * x2 + m3 * x3 + m4 * x4 + m5 * x5 + c
y_pred_2
```

So, in both cases prediction is same

# r2_score

```python
score = r2_score(y, y_pred)
score
```
