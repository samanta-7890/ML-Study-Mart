# Ordinal Encoder 

```python
df7.head()
```

On Sex Column

```python
df7['sex'].unique()
```

```python
sex_unique = ['female', 'male']
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[sex_unique])
```

```python
df7['sex'] = oe.fit_transform(df7[['sex']])
df7.head()
```

On smoker Column

```python
smoke_unique = df7['smoker'].unique()
```

```python
oe2 = OrdinalEncoder(categories=[smoke_unique])
df7['smoker'] = oe2.fit_transform(df7[['smoker']])
df7.head()
```

On region Column

```python
region_unique = df7['region'].unique()
```

```python
oe3 = OrdinalEncoder(categories=[region_unique])
df7['region'] = oe3.fit_transform(df7[['region']])
df7.head()
```

# Ordinal Encoder Using For Loop  

```python
df8.head()
```

```python
for col in df8.columns:
    if is_numeric_dtype(df8[col]):
        continue
    else:
        oe = OrdinalEncoder(categories = [df8[col].unique()])
        df8[col] = oe.fit_transform(df8[[col]])
```
